# Podaci

```python
M = np.zeros([50, 2])
M[:, 0] = np.random.uniform(10, 20, 50)
M[:, 1] = 3*np.exp(0.3 * M[:, 0]) + np.random.normal(20, 50, 50) 
```

```python
X = M[:, 0]
Y = M[:, 1]
```

```python
plt.figure()
plt.plot(X, Y, 'o')
```

## Regresija eksponencijalnog modela
Model:
$$y = a e^{bx}$$
Moe se zapisati kao
$$ \ln y = \ln a + bx$$

Odnosno, provodimo linearnu regresiju za podatke **$\ln y$** i **$x$**. 

Rezultat su koeficijenti pravca $\ln a$ i $b$



```python
plt.figure()
plt.plot(X, Y, 'o')
x = np.linspace(X.min(), X.max(), 200)

k = np.polyfit(X, np.log(Y), 1)

print(f'Dobiveni koeficijenti:\nln(a)={k[1]:.4f}\nb={k[0]:.4f}')
print(f'Sada je a={np.exp(k[1]):.4f}')
a = np.exp(k[1])
b = k[0]
print(f'Model je: {a:.4f}*e**(x*{b:.4f})')
# print(f'{k3}, {np.exp(k3)}')
plt.plot(x, a * np.exp(x * b), label='Eksponencijalna regresija')
# plt.legend()
draw_grids()
```
